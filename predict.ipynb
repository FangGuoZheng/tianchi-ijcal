```python
import pickle as pk
import pandas as pd

# START CODE HERE ###
train = open('../data/train01.pkl', 'rb')
train = pk.load(train)
train.head(6)
# train.dtypes
# END CODE HERE ###
```

```python
train = train.loc[train["context_timestamp"] < '2018-09-24 00:00:00']
train["context_timestamp"].tail()
```

```python
train1 = train.groupby(train['shop_id'])
train1_grp = train1.size()
train1_grp = train1_grp.astype('float32')
train1_grp.head(6)
print(len(train1_grp))
# train1_grp.dtypes
# type(train1_grp)

train_temp = train.loc[train["is_trade"] == '1']
train2 = train_temp.groupby(train_temp['shop_id'])
train2_grp = train2.size()
train2_grp = train2_grp.astype('float32')
print(len(train2_grp))
for k, v in train1_grp.items():
    if k in train2_grp.index:
        train1_grp[k] = train2_grp[k]/v
    else:
        train1_grp[k] = 0

count = 0
for k, v in train1_grp.items():
    if v == 0.0:
        count=count+1
print(count)
# len(train1_grp)
train1_grp.head(6)
```

```python
train3_grp = pd.DataFrame({'shop_id' : pd.Series(dtype='uint64'), 'purchase_rate' : pd.Series(dtype='float32')})
train3_grp['shop_id'] = train1_grp.index
train3_grp['purchase_rate'] = train1_grp.values
# 类型转换
# train3_grp["purchase_rate"] = train3_grp["purchase_rate"].astype('float32')
train3_grp.head(6)
# len(train3_grp)
# train3_grp.dtypes
# type(train3_grp)
```
